Ising machine: Job Sequencing

you have N jobs and M computer clusters. Each job i has a length L_i. You want
to assign every job to a cluster such that the longest cluster finishes as
quickly as possible. This has applications to load balancing

spin vector: spins[m][n] = 1 if job n is in cluster m
auxillary spin vector y: y_{n,α} auxiliary spins encode the difference M₁ - M_α
= n for each other cluster α.

Hamiltonian Penalty Terms: 

1. Enforces each job is only assigned to one cluster. Also enforces cluster 1 as
   longest running (without loss of generality)
2. minimizes duration of longest cluster.
