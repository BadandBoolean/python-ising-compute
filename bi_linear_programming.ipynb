Ising machine: Binary Integer Linear Programming

You have N binary variables x₁, ..., x_N, and you want to maximize c · x subject
to Sx = b, where S is an m×N matrix and b is a vector with m components.

Example: A factory makes 3 products. Each product is either made (1) or not (0).
You want to maximize profit subject to resource constraints.

Hamiltonian penalty terms:

1. Enforces constraint Sx = b
2. maximuse objective cx for some vector c. 

NP-hard problem. ILP is generalization (but not necessarily simplification) of
other Karp NP problems.
